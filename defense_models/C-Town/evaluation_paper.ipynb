```python
from aeed_utils import *
import matplotlib.pyplot as plt
%matplotlib inline
from numpy.random import seed
seed(1)
import tensorflow
tensorflow.random.set_seed(1)
```

```python
### Load and preprocess training data
data_path = '../../dataset/normal_operating_conditions_3_0/output/'
file_expr = data_path + '*'
paths = glob.glob(file_expr)
```

```python
physical_pd = None
for path in paths:
    aux = preprocess_physical(path)
    physical_pd = pd.concat([physical_pd, aux])

if'Unnamed: 0' in physical_pd.columns:
    physical_pd = physical_pd.drop(columns=['Unnamed: 0'])
if 'ATTACK' in physical_pd.columns:
    physical_pd = physical_pd.drop(columns=['ATTACK'])
physical_pd.to_csv('training_data.csv',index=False)
```

```python
dates_train = physical_pd['timestamp']
sensor_cols = [col for col in physical_pd.columns if col not in ['Unnamed: 0','iteration', 'timestamp', 'Attack']]
print(sensor_cols)
```

```python
scaler = MinMaxScaler()
X = scaler.fit_transform(physical_pd[sensor_cols])

# define model parameters
params = {
    'nI' : X.shape[1],
    'nH' : 2,
    'cf' : 2.5,
    'activation' : 'tanh',
    'verbose' : 1,
}

# create AutoEncoder for Event Detection (AEED)
autoencoder = AEED(**params)
autoencoder.initialize()


X1, X2, _, _  = train_test_split(X, X, test_size=0.33, random_state=42)
```

```python
# # train models with early stopping and reduction of learning rate on plateau

earlyStopping= EarlyStopping(monitor='val_loss', patience=3, verbose=0,  min_delta=1e-4, mode='auto')
lr_reduced = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=1, verbose=0, epsilon=1e-4, mode='min')

# initialize time
start_time = time.time()

# train autoencoder
training_object = autoencoder.train(X1, X1,
            epochs=500,
            batch_size=32,
            shuffle=True,
            callbacks = [earlyStopping, lr_reduced],
            verbose = 2,
            validation_data=(X2, X2))  

_, validation_errors = autoencoder.predict(X2, X2)

end_time = time.time()
```

```python
# get validation reconstruction errors
_, validation_errors = autoencoder.predict(X2, X2)


# plot distribution of average validation reconstruction errors 
f, ax = plt.subplots(1, figsize = (8,4))
sns.boxplot(validation_errors.mean(axis=1), ax=ax)
ax.set_title('Boxplot of average validation reconstruction errors')

# set treshold as quantile of average reconstruction error
theta = validation_errors.mean(axis = 1).quantile(0.9722222222222222)
print(theta)
```

```python
import yaml
def find_attack_intervals(path):
    start = np.inf
    end = -np.inf
    with open(path) as f:
        my_dict = yaml.safe_load(f)
#         print(my_dict)
        try:
            for i in range(0,len(my_dict['attacks']['device_attacks'])):
                start =  my_dict['attacks']['device_attacks'][i]['trigger']['start'] if start > my_dict['attacks']['device_attacks'][i]['trigger']['start'] else start
                end = my_dict['attacks']['device_attacks'][i]['trigger']['end'] if end < my_dict['attacks']['device_attacks'][i]['trigger']['end'] else end
        except KeyError:
            for i in range(0,len(my_dict['attacks']['network_attacks'])):
                start =  my_dict['attacks']['network_attacks'][i]['trigger']['start'] if start > my_dict['attacks']['network_attacks'][i]['trigger']['start'] else start
                end = my_dict['attacks']['network_attacks'][i]['trigger']['end'] if end < my_dict['attacks']['network_attacks'][i]['trigger']['end'] else end
    return start, end+1

def add_gt(dataset, attack_start, attack_end):
    array = np.array([0]*len(dataset))
    array[dataset['iteration'][dataset['iteration']== attack_start].index[0]:dataset['iteration'][dataset['iteration']== attack_end].index[0]] += 1
    dataset['ATTACK'] =  array
    return dataset
```

```python
def load_conceament_data(index):
    preamble_folder = '../../dataset/evasion_data/'
    attack  = pd.read_csv(preamble_folder+'attack_output_'+str(index)+'/scada_values.csv')
    if index == 35:
        attack = attack.drop(attack[attack['iteration'].duplicated()].index)
    file = preamble_folder+'attack_output_'+str(index)+'/configuration/config.yaml'
    attack_start, attack_end = find_attack_intervals(file)
    attack  = add_gt(attack, attack_start, attack_end)
    return attack
```

```python
attacks = []
attack_names = [] 
for i in range(1,38):
    if i < 10:
        i = format(i, '02d')
    try:
        attacks.append(load_conceament_data(i))
        attack_names = attack_names + ['attack_'+str(i)]
    except:
        continue
        
attacks.append(load_conceament_data(44))
attack_names = attack_names + ['attack_'+str(44)]
```

```python
def mapping(attack_name):
    if any(x in attack_name for x in ['01' , '02' , '44']):
        return 'Anomaly 1'
    if any(x in attack_name for x in ['03', '04' , '05']):
        return 'Anomaly 1.2'
    if any(x in attack_name for x in ['06' , '07' ]):
        return 'Anomaly 2'
    if any(x in attack_name for x in [ '08' , '09' , '10']):
        return 'Anomaly 2.2'
    if any(x in attack_name for x in ['11' , '12' , '13']):
        return 'Anomaly 3'
    if any(x in attack_name for x in ['14' , '15']):
        return 'Anomaly 4'
    if any(x in attack_name for x in ['25' , '26']):
        return 'Anomaly 9'
    if any(x in attack_name for x in ['27' , '28' , '29']):
        return 'Anomaly 12'
    if any(x in attack_name for x in ['34' , '35' ]):
        return 'Anomaly 14'
    if any(x in attack_name for x in ['36', '37']):
        return 'Anomaly 14.2'
```

```python
results_y = []
results_y_hat = []
gt_list = [Y3,Y3,Y3,Y3,Y9,Y3,Y3,Y3,Y3,Y3,Y4,Y4,Y4,Y5,Y5,Y6,Y6,Y7,Y7,Y7,Y8,Y8,Y8,Y8,Y9]
for attack, attack_name,gt in zip(attacks, attack_names,gt_list):
    print(attack_name)
    
    attack_data = pd.DataFrame(index = attack.index, columns = sensor_cols, 
                      data = scaler.transform(attack[sensor_cols]))
    prediction, _ = autoencoder.detect(attack_data, attack_data, theta = theta , window = a_window, average=True)
    anomaly_name = mapping(attack_name)
    results.loc[anomaly_name+ ' ' + attack_name] = compute_scores(gt, prediction, a_window, theta)
    results_y = results_y + [gt]
    results_y_hat = results_y_hat + [prediction]
```
